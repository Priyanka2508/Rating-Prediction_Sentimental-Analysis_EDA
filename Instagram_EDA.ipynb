```python
#importing

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn-pastel')
```

```python
post=list(df['Posts'])

post_new = []

for num in post:
  num = 1000*float(num.replace('k',''))
  post_new.append(num)

post_new
```

```python
df['Posts'] = post_new
df['Posts']
```

```python
followers1 = list(df['Followers'])
followers_new=[]
for each in followers1:
    if 'm' in str(each):
        each = 1000000 * float(str(each).replace('m', ''))
        followers_new.append(each)
    else:
        each=float(each)
        followers_new.append(each)
df['Followers']=followers_new


avg_likes=list(df['Avg. Likes'])
avg_likes1=[]
for each in avg_likes: #In Avg likes column, there are million(m),Thousand(k)
    if 'm' in str(each):
        each=1000000*float(str(each).replace('m',''))
        avg_likes1.append(each)

    elif 'k' in str(each):
        each=1000*float(str(each).replace('k',''))
        avg_likes1.append(each)
    else:
        each=float(str(each))
        avg_likes1.append(each)
df['Avg. Likes']=avg_likes1

new_post_avg_like=list(df['New Post Avg. Likes'])
new_post_avg_like1=[]
for each in new_post_avg_like:
    if 'm' in str(each):
        each=1000000*float(str(each).replace('m',''))
        new_post_avg_like1.append(each)

    else:
        each=1000*float(str(each).replace('k',''))
        new_post_avg_like1.append(each)
df['New Post Avg. Likes']=new_post_avg_like1



total_likes=list(df['Total Likes'])
total_likes1=[]
for each in total_likes:
    if 'm' in str(each):
        each=1000000*float(str(each).replace('m',''))
        total_likes1.append(each)

    elif 'b' in str(each):
        each=1000000000*float(str(each).replace('b',''))
        total_likes1.append(each)

df['Total Likes']=total_likes1
```

# **5. Relationship between the following pair of features**

```python
from scipy.stats import linregress
```

**5.1 Followers and Total Likes**

```python
plt.scatter(df['Followers'], df['Total Likes'])
plt.title("Relationship between Followers and Total Likes")
plt.xlabel("Followers")
plt.ylabel("Total Likes")

sns.regplot(x='Followers', y='Total Likes', data=df, scatter_kws={'s': 100},line_kws={'color':'red'})
plt.title('Scatter Plot with Regression Line')
plt.xlabel('Followers')
plt.ylabel('Total Likes')

#  Calculate correlation coefficient
correlation_coefficient = df['Followers'].corr(df['Total Likes'])
print(f'Correlation Coefficient: {correlation_coefficient:.2f}')

# Calculate equation of the regression line
slope, intercept, r_value, p_value, std_err = linregress(df['Followers'], df['Total Likes'])
equation = f'Total Likes = {slope:.2f} * Followers + {intercept:.2f}'
plt.annotate(equation, xy=(0.5, 0.9), xycoords='axes fraction', fontsize=10, color='black')

plt.show()
```

**5.2 Followers and Influence Score**

```python
plt.scatter(df['Followers'], df['Influence Score'])
plt.title("Relationship between Followers and Influence Score")
plt.xlabel("Followers")
plt.ylabel("Influence Score")

sns.regplot(x='Followers', y='Influence Score', data=df, scatter_kws={'s': 100},line_kws={'color':'red'})
plt.title('Scatter Plot with Regression Line')
plt.xlabel('Followers')
plt.ylabel('Influence Score')

#  Calculate correlation coefficient
correlation_coefficient = df['Followers'].corr(df['Influence Score'])
print(f'Correlation Coefficient: {correlation_coefficient:.2f}')

# Calculate equation of the regression line
slope, intercept, r_value, p_value, std_err = linregress(df['Followers'], df['Influence Score'])
equation = f'Influence Score = {slope:.2f} * Followers + {intercept:.2f}'
plt.annotate(equation, xy=(0.5, 0.9), xycoords='axes fraction', fontsize=10, color='black')

plt.show()
```
